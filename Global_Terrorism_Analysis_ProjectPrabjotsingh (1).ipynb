# Cleaning the data

###  Note: Since it contains 135 columns. They have a huge proportion in dataset and Learning them doesn't make any sense. So, we will rename the columns name for better understaning and then we will only extract necessary columns.

#### Correct feature name

```python
global_terror=global_terror[["iyear","imonth","iday","country_txt","region_txt","provstate","city",
       "latitude","longitude","location","summary","attacktype1_txt","targtype1_txt",
       "gname","motive","weaptype1_txt","nkill","nwound","addnotes"]]
global_terror.head()
```

### Important Data for the further processing

```python
global_terror.rename(columns={"iyear":"Year","imonth":"Month","iday":"Day","country_txt":"Country",
                   "region_txt":"Region","provstate":"Province/State","city":"City",
                   "latitude":"Latitude","longitude":"Longitude","location":"Location",
                   "summary":"Summary","attacktype1_txt":"Attack Type","targtype1_txt":"Target Type",
                   "gname":"Group Name","motive":"Motive","weaptype1_txt":"Weapon Type",
                   "nkill":"Killed","nwound":"Wounded","addnotes":"Add Notes"},inplace=True)
global_terror.head()
```

```python
global_terror.shape
```

```python
global_terror.isnull().sum()
```

### Remove null value

```python
global_terror["Killed"]=global_terror["Killed"].fillna(0)
global_terror["Wounded"]=global_terror["Wounded"].fillna(0)
global_terror["Consualty"]=global_terror["Killed"]+global_terror["Wounded"]
```

```python
global_terror.describe()
```

### 1. Year wise Attacks

```python
attacks=global_terror["Year"].value_counts(dropna=False).sort_index().to_frame().reset_index().rename(columns={"index":"Year","Year":"Attacks"}).set_index("Year")
attacks.head()
```

## 2. Region wise Attacks

### 1. Distribution of Terrorist Attacks over Regions from 1970-2017

```python
reg=pd.crosstab(global_terror.Year,global_terror.Region)
reg.head()
```

```python
reg.plot(kind="area", stacked=False, alpha=0.5,figsize=(20,10))
plt.title("Region wise attacks",fontsize=20)
plt.xlabel("Years",fontsize=20)
plt.ylabel("Number of Attacks",fontsize=20)
plt.show()
```

### 2. Total Terrorist Attacks in each Region from 1970-2017

```python
regt=reg.transpose()
regt["Total"]=regt.sum(axis=1)
ra=regt["Total"].sort_values(ascending=False)
ra
```

```python
ra.plot(kind="bar",figsize=(15,6))
plt.title("Total Number of Attacks in each Region from 1970-2017")
plt.xlabel("Region")
plt.ylabel("Number of Attacks")
plt.show()
```
